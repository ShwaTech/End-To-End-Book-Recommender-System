#### Feature Engineering ❤️‍🔥❤️‍🔥

```python
# Lets store users who had at least rated more than 200 books

X = ratings["user_id"].value_counts() > 200

X
```

```python
X[X].shape
```

```python
y = X[X].index

y
```

```python
ratings = ratings[ratings["user_id"].isin(y)]

ratings.head()
```

```python
# Now join ratings with books

ratings_with_books = ratings.merge(books, on="ISBN")

ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby("title")["rating"].count().reset_index()

number_rating.head()
```

```python
number_rating.rename(columns={"rating": "num_of_rating"}, inplace=True)

number_rating.head()
```

```python
final_rating = ratings_with_books.merge(number_rating, on="title")

final_rating.head()
```

```python
# Lets take those books which got at least 50 rating of user

final_rating = final_rating[final_rating['num_of_rating'] >= 50]

final_rating.head()
```

```python
# Let's Drop the Duplicates

final_rating.drop_duplicates(["user_id", "title"], inplace=True)
```

```python
final_rating.isnull().sum()
```

```python
# Lets create a pivot table

books_pivot = final_rating.pivot_table(columns="user_id", index="title", values="rating")

books_pivot
```

```python
books_pivot.shape
```

```python
books_pivot.fillna(0, inplace=True)
```

```python
books_pivot
```

Conver book_povit To CSR (Compressed Sparse Row) Matrix.
- In CSR, only non-zero elements are stored along with their row indices and column indices.
- This format enables efficient arithmetic operations, row slicing, and quick access to non-zero elements, making it ideal for sparse data.

```python
from scipy.sparse import csr_matrix

books_sparse = csr_matrix(books_pivot)

books_sparse
```

```python
type(books_sparse)
```

#### Model Training ??

We Start The Model Using Unsupervised Popular Technique **NearestNeighbors** (It's Not Like Supervised KNearestNeighbors)
Take Care Not To Conflict

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(algorithm="brute")
```

```python
model.fit(books_sparse)
```

```python
books_pivot.iloc[237, :]
```

```python
# Use kneighbors Function To Get The N Nearst Neighbors (Have The Similar Rating - Similar Interest)

distance, suggestion = model.kneighbors(books_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors=6)
```

#### Testing Model ??

```python
def recommend_book(book_name):
    book_id = np.where(books_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(books_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)
    for i in range(len(suggestion)):
        books = books_pivot.index[suggestion[i]]
        for j in books:
            if j == book_name:
                print(f"You Searched '{book_name}'\n")
                print("The suggestion books are: \n")
            else:
                print(j)
```

```python
book_name = "2nd Chance"
recommend_book(book_name)
```
